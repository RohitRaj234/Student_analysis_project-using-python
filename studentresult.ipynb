```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#drop unnamed column
df=df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
#gender distribution
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#from the above chart we have analysis that:
#the number of female in the data is more than  the number of females

```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Edu and student's score")
plt.show()
```

#from the above chart we have concluded that the education of parents have good impact

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent's Martial status and student's score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible impact on the student's score due to their parent's martial status
```
